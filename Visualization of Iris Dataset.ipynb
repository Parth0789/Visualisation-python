```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris()
```

```python
df = pd.DataFrame(data= np.c_[df['data'], df['target']],
                     columns= df['feature_names'] + ['target'])
```

```python
df.head()
```

```python
df['sepal length (cm)'].plot(kind='line')
```

```python
plt.subplot(1,2,1) #plt.subplot(row,columns, position) 
df['sepal length (cm)'].plot(kind='line')
plt.subplot(1,2,2)
df['petal length (cm)'].plot(kind='line')
```

```python
plt.subplot(2,1,1) #plt.subplot(row,columns, position) 
df['sepal length (cm)'].plot(kind='line')
plt.subplot(2,1,2)
df['petal length (cm)'].plot(kind='line')
```

<h2>Compare two variable(Use Area graph) 

```python
df[['petal length (cm)', 'sepal length (cm)']].plot(kind='area')
```

<h2>Histogram

```python
df['petal length (cm)'].plot(kind='hist')
```

```python
df.plot(kind='scatter', x='petal length (cm)', y='petal width (cm)')
```
